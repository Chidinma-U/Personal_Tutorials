### Filtering Data from Dataframes and series objects

```python
#trying to make a data for meeting who are making a salary over a certain amount
high_salary = (df['ConvertedComp']>70000)
```

```python
df.loc[high_salary]
```

```python
#filtering out the country, programming languages and the salary
df.loc[high_salary, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
#to filter out by different countries

countries = ['United States', 'Germany', 'India', 'United Kingdom', 'Canada']
filt = df['Country'].isin(countries)
```

```python
df.loc[filt]
```

```python
#we can also use string method to do some alterations to our data frame or to help with a conditional

#to get prople who use 'Python' programming language
filt = df['LanguageWorkedWith'].str.contains('Python', na = False) #does the column 'languageworked with' contain 'Python'?
```

```python
df.loc[filt, 'LanguageWorkedWith']
```

## How to alter existing rows and columns in our dataframe

```python
df['SalaryUSD']
```

We want to convert all the 'Yes/No' values in this column to 'True/False' values.

```python
df['Hobbyist'].map({'Yes':True, 'No':False}).head()
#Remember that '.head()' prints just the first 5 rows in your column
```

## How to Sort Our Data

#### Sorting by country name

```python
df = pd.read_csv('data/survey_results_public.csv')
```

```python
df.sort_values(by='Country', inplace = True)
```

```python
df[['Country', 'ConvertedComp']].head(50)
```

```python
#To see the country in ascending order, but the salary in descending order (from the highest to the lowest)

df.sort_values(by=['Country', 'ConvertedComp'], ascending = [True, False], inplace = True)
```

```python
df[['Country', 'ConvertedComp']].head(50)
```

Grabbing the largest or smaller values from a specific dataframe

```python
df['ConvertedComp'].nlargest(10) #To see those who have the highest salary.
```

```python
df.nlargest(10, 'ConvertedComp')
```
